```python
df2=df[['PCOS (Y/N)', ' Age (yrs)', 'Weight (Kg)','Height(Cm) ', 'BMI', 'Blood Group', 'Pulse rate(bpm) ','Cycle(R/I)', 'Cycle length(days)','Pregnant(Y/N)', 'FSH(mIU/mL)',
       'LH(mIU/mL)', 'FSH/LH','Weight gain(Y/N)', 'hair growth(Y/N)','Skin darkening (Y/N)', 'Hair loss(Y/N)', 'Pimples(Y/N)','Reg.Exercise(Y/N)','Endometrium (mm)','Follicle No. (L)', 'Follicle No. (R)']]
```

```python
df2.head()
```

```python
df2.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df2.drop('PCOS (Y/N)',axis=1)
y=df2['PCOS (Y/N)']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
prediction=rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(prediction,y_test))
```

```python
confusion_matrix(prediction,y_test)
```

```python
from joblib import Parallel, delayed
import joblib
joblib.dump(rfc, 'PCOS.pkl')
  
```
